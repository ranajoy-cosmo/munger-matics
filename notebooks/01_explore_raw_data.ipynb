```python
from pathlib import Path
import polars as pl
```

## 1. Locate raw files

```python
raw_dir = Path("../data/raw")
files = list(raw_dir.iterdir())
files
```

## 2. Inspect the raw file

```python
# Set the file to explore
raw_file = files[0]

# Read first few lines as plain text to check encoding and structure
with open(raw_file, encoding="utf-8") as f:
    for i, line in enumerate(f):
        print(repr(line))
        if i >= 9:
            break
```

## 3. Load with Polars

```python
# Adjust separator, skip_rows, encoding as needed after inspecting above
df = pl.read_csv(raw_file, separator=";")
df.head(10)
```

```python
# Schema — column names and inferred types
print(df.schema)
```

```python
# Shape and basic stats
print(f"Rows: {df.height}, Columns: {df.width}")
df.describe()
```
